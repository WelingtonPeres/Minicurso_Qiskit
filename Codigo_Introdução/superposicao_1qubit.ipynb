```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator # Nosso simulador quântico

# Construção do circuito quântico
qc = QuantumCircuit(1, 1)

qc.h(0) # Aplica a porta Hadamard ao qubit 0

qc.measure(0, 0) # Mede o qubit 0 e armazena no bit clássico 0

display(qc.draw('mpl')) 
```

```python
# Simulação do circuito quântico
simulator = AerSimulator()

job = simulator.run(qc, shots=1024)

result = job.result()
counts = result.get_counts(qc)

print("\nContagem dos resultados (0s e 1s):", counts)
```

```python
plot_histogram(counts)
```
