```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
# Qual tamanho do cofre (número de qubits)
n = 3

initial_state = '101'
```

```python
# Iniciando o circuito quântico

qc_cofre = QuantumCircuit(n, n)

qc_cofre.x(0)
qc_cofre.x(2)
qc_cofre.barrier(label="Setup")

qc_cofre.cx(0, 1)
qc_cofre.ccx(0, 1, 2)
qc_cofre.x(1)
qc_cofre.barrier(label="Cofre")

qc_cofre.measure(range(n), range(n))

display(qc_cofre.draw('mpl'))
```

```python

simulator = AerSimulator()
job = simulator.run(qc_cofre, shots=1024)
counts = job.result().get_counts()

plot_histogram(counts)
```
